```python
#Change this before running
WeekTrack = "Week3"


DraftFile = WeekTrack + "/" + "Draft_List_Start_" + WeekTrack + ".csv"
RequestFile = WeekTrack + "/"+"Request_" + WeekTrack + ".csv" 
FinalPicks = WeekTrack + "/" + "Final_picks_" + WeekTrack +".csv"
AddFile = WeekTrack + "/" + "AddDrop_" + WeekTrack + ".csv"
RevisedDraftFile = WeekTrack + "/" +"Draft_List_End" + WeekTrack + ".csv"


```

```python
import pandas as pd
import numpy as np

Draftpick_DF = pd.read_csv(DraftFile, skiprows= np.arange(31,240,31))

Draftpick_DF.head(65)
```

```python
PickList = []

for draft in Draftpick_DF["Round/Pick"]:
    splitpick = draft.split("/")[1].lstrip()
    splitpick = int(splitpick)
    PickList.append(splitpick)

Draftpick_DF["Pick"] = PickList

Draftpick_DF.head(50)

    
    
```

```python
RoundList = []

for draft in Draftpick_DF["Round/Pick"]:
    splitround = draft.split("/")[0].rstrip()
    splitround = int(splitround)
    RoundList.append(splitround)
    
Draftpick_DF["Round"] = RoundList

Draftpick_DF.head(50)
```
